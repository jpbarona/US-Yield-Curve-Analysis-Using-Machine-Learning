```python
import pandas as pd

raw = pd.read_csv("USCurves.csv")

MONTH = "m"
YEAR = "y"
raw.columns
```

```python
column_mutations = {}
```

```python
i = 1
x = raw.columns[i]
x
```

```python

duration_raw = None
for char in x:
    try:
        duration_raw = int(char)
    except ValueError:
        continue

if duration_raw is None:
    raise ValueError(f"Expected number in string {x!r}")

period_identifier = x.split(str(duration_raw))[1].lower()[0]
year_duration = -1
if period_identifier == MONTH:
    year_duration = float(duration_raw / 12)
else:
    year_duration = float(duration_raw)

column_mutations[raw.columns[i]] = year_duration
```

```python
for i in range(len(raw.columns)-1):
    x = raw.columns[i]
    if "BC" not in x:
        continue

    duration_raw = None
    for char in x:
        try:
            duration_raw = int(char)
        except ValueError:
            continue    

    if duration_raw is None:
        raise ValueError(f"Expected number in string {x!r}")

    period_identifier = x.split(str(duration_raw))[1].lower()[0]
    year_duration = -1
    if period_identifier == MONTH:
        year_duration = float(duration_raw / 12)
    else:
        year_duration = float(duration_raw)
    
    column_mutations[raw.columns[i]] = year_duration

    
```
